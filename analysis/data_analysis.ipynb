### duration_sa2_hospital_shortest.pkl
#### Dataset:

```python
data = pd.read_pickle(os.path.join(datadir, "duration_sa2_hospital_shortest.pkl"))
data.head()
```

#### Notes:
This is the shortest travel time between Statistical Areas Level 2 (SA2s) and any hospital in
Australia. However, the hospital locations are not provided in this dataset, only the travel times.

SA2 code column format: 11007 (SA2_5DIG16)

NOTE: This is also available for the Mesh Block level, which is a more granular level than SA2.

### duration_mb_hospitals.pkl

## Additional analysis
### Hospital metadata
Join the hospital metadata datasets to get a complete list of hospitals in Australia. See if rows are lost in the process.

```python
hospital_details = pd.read_csv(os.path.join(datadir, "myhospitals-contact-details.csv"), encoding="latin1")
hospital_list = pd.read_csv(os.path.join(datadir, "public_hospital_list.csv"))

# Join the two tables
hospital_data = hospital_details.merge(hospital_list, on="Hospital name")
hospital_data.head()
```

#### Notes:
Nice! The hospital metadata datasets have different columns, so we can join them to get a complete list of hospitals in Australia. (e.g. first dataset has `Beds` column, second dataset has `Number of available beds` column, this can be useful for further analysis)
