```python
import pandas as pd
```

```python
file_path = r"C:\Users\wfeij\Downloads\Mesh Block Counts 2021.xlsx"

# Load the Excel file
xls = pd.ExcelFile(file_path)

# List to hold DataFrames for each sheet
sheet_list = []

# Iterate through each sheet and read it into a DataFrame
for sheet_name in xls.sheet_names[2:]:
    df = pd.read_excel(xls, sheet_name=sheet_name,skiprows=6)
    sheet_list.append(df)

# Concatenate all DataFrames
concatenated_df = pd.concat(sheet_list, ignore_index=True)

# concatenated_df[['MB_CODE_2021', 'MB_CATEGORY_NAME_2021', 'AREA_ALBERS_SQKM', 'Dwelling', 'Person', 'State','Geographic Extent Name']].to_pickle("new data/mb_data_2021.pkl")
```
